```python
import pandas as pd

df = pd.read_csv("gpu_workload_data.csv")

print(df)
```

```python
def detect_stress(row):

    load = row["Load (%)"]
    temp = row["Temperature (C)"]

    if load > 80 or temp > 80:
        return "High Stress"

    elif load > 50 or temp > 70:
        return "Medium Stress"

    else:
        return "Low Stress"

df["Stress_Level"] = df.apply(detect_stress, axis=1)

print(df)
```

```python
print(df["Stress_Level"].value_counts())
```

```python
import matplotlib.pyplot as plt

plt.plot(df["Load (%)"])
plt.plot(df["Temperature (C)"])

plt.xlabel("Time Index")
plt.ylabel("Value")
plt.title("GPU Stress Monitoring")

plt.legend(["GPU Load","Temperature"])

plt.show()
```
